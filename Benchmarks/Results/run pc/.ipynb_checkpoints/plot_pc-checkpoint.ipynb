```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
opt = []
naive = []
for i in range(4):
    opt.append(pd.read_csv("opt"+str(i+1)+".txt",sep='\t',names=['i', 'time','hashrate']))
    naive.append(pd.read_csv("naive"+str(i+1)+".txt",sep='\t',names=['i', 'time','hashrate']))
```

```python

for i in range(4):
    opt[i]['n'] = 2**opt[i]['i']
    naive[i]['n'] = 2**naive[i]['i']
    
```

```python
for i in range(4):
    df = opt[i]
    opt[i]['hashrate'] = pd.to_numeric(df[df['hashrate'].apply(lambda x: x != "1.#INF00")]['hashrate'],downcast='float')
    df = naive[i]
    naive[i]['hashrate'] = pd.to_numeric(df[df['hashrate'].apply(lambda x: x != "1.#INF00")]['hashrate'],downcast='float')
```

```python
for i in range(1,4):
     opt[i] = opt[i].drop(['i', 'n'], axis=1)
final_opt = opt[0]
for i in range(1,4):
    final_opt = pd.concat([final_opt,opt[i]],axis=1)
```

```python
for i in range(1,4):
     naive[i] = naive[i].drop(['i', 'n'], axis=1)
final_naive = naive[0]
for i in range(1,4):
    final_naive = pd.concat([final_naive,naive[i]],axis=1)
```

```python
final_opt
```

```python

hash_row_opt = []
for rows in final_opt.itertuples():
    count = 0
    sums = 0
    if not np.isnan(rows[3]):
        sums += rows[3]
        count += 1
    if not np.isnan(rows[6]):
        sums += rows[6]
        count += 1
    if not np.isnan(rows[8]):
        sums += rows[8]
        count += 1
    if not np.isnan(rows[10]):
        sums += rows[10]
        count += 1
    if count == 0:
        avg = np.nan
    else:
        avg = sums/count
    hash_row_opt.append(avg)
final_opt['avg_hash'] = hash_row_opt
```
